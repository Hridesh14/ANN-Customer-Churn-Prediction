```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
## load ann trained model sclr pkl ohe lab 
model = load_model('model.h5')
```

```python
with open('lable_encoder_gender.pkl','rb') as file1:
    laber_gender = pickle.load(file1)

with open('OneHotEncoder_geo.pkl','rb')as file2:
    one_hot_geo = pickle.load(file2)

with open('scaler.pkl','rb') as file3:
    scaled_opp = pickle.load(file3)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
input_df = pd.DataFrame([input_data])
```

```python
geo_encoder = one_hot_geo.transform([[input_data['Geography']]]).toarray()
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder, columns=one_hot_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
input_df = pd.concat([input_df.reset_index(drop=True),geo_encoded_df],axis=1)
```

```python
input_df
```

```python
input_df['Gender']=laber_gender.transform(input_df['Gender'])
```

```python
input_df
```

```python
input_df.drop('Geography',axis=1,inplace=True)
```

```python
scalling = scaled_opp.transform(input_df)
```

```python
scalling
```

```python
## prediction 
pred =model.predict(scalling)
```

```python
prediction_prob = pred[0][0]
```

```python
prediction_prob
```

```python
if prediction_prob > 0.5:
    print('The Customer is likely to churn')
else:
    print('The Customer is not likely to churn')
```
